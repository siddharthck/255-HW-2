```python
import matplotlib.pyplot as plt

methods = [
    "sklearn LR", "Intel LR", "RAPIDS LR",
    "sklearn RF", "Intel RF", "RAPIDS RF",
    "PyTorch NN"
]
accuracies = [
    0.8659, 0.8659, 0.8658,
    0.8671, 0.8671, 0.8662,
    0.8649
]

plt.figure(figsize=(10, 5))
plt.bar(methods, accuracies, color="slateblue")
plt.title("Accuracy Comparison: LR, RF, and PyTorch NN")
plt.ylabel("Accuracy")
plt.ylim(0.86, 0.87)
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.savefig("all_accuracy_comparison.png", dpi=300)  # Save here

plt.show()
```

```python
train_times = [
    0.1131, 0.1142, 0.1990,     # LR
    4.1445, 3.6874, 2.3145,     # RF
    112.4355                    # NN
]

plt.figure(figsize=(10, 5))
plt.bar(methods, train_times, color="darkorange")
plt.yscale("log")  # Use log scale for better visibility
plt.title("Training Time Comparison (Log Scale): LR, RF, and PyTorch NN")
plt.ylabel("Training Time (log seconds)")
plt.xticks(rotation=45)
plt.grid(axis="y", which='both', linestyle='--')
plt.tight_layout()
plt.savefig("all_training_comparison.png", dpi=300)  # Save here
plt.show()

```
